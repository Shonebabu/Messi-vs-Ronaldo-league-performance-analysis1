```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lm10cr7=pd.read_csv('MessiRonaldo.csv')
```

```python
lm10cr7.head()
```

```python
lm10cr7.columns
```

```python
lm10cr7.info()
```

```python
lm10cr7['Mins/Liga_Goals'] = lm10cr7['Liga_Mins']/lm10cr7['Liga_Goals']
lm10cr7['Mins/Liga_Asts'] = lm10cr7['Liga_Mins']/lm10cr7['Liga_Asts']
```

```python
lm10cr7.head()
```

```python
lm10cr7['Mins/Liga_Aps'] = lm10cr7['Liga_Mins']/lm10cr7['Liga_Aps']
lm10cr7['Mins/CL_Aps'] = lm10cr7['CL_Mins']/lm10cr7['CL_Aps']
lm10cr7['Mins/CL_Goals'] = lm10cr7['CL_Mins']/lm10cr7['CL_Aps']
lm10cr7['Mins/CL_Asts'] = lm10cr7['CL_Mins']/lm10cr7['CL_Asts']
```

```python
lm10cr7.head()
```

```python
lm10cr7['lg_rank'] = [1,3,2,1,1,2,1,2,2,3,1,2,2,1,1,2,1,2]
lm10cr7['CL_rank'] = [5,1,5,1,5,1,1,4,5,1,4,4,4,4,1,4,4,5]
```

```python
plt.figure(figsize=(15,15))
plt.subplot(421)
lm10cr7.groupby('Player')['Liga_Goals'].sum().plot.bar()
plt.title('Liga_Goals')


plt.subplot(422)
lm10cr7.groupby('Player')['Liga_Asts'].sum().plot.bar()
plt.title('Liga_Asts')

plt.subplot(423)
lm10cr7.groupby('Player')['Liga_Aps'].sum().plot.bar()
plt.title('Liga_Aps')

plt.subplot(424)
lm10cr7.groupby('Player')['CL_Goals'].sum().plot.bar()
plt.title('CL_Goals')

plt.subplot(425)
lm10cr7.groupby('Player')['CL_Asts'].sum().plot.bar()
plt.title('CL_Asts')

plt.subplot(426)
lm10cr7.groupby('Player')['CL_Aps'].sum().plot.bar()
plt.title('CL_Aps')

plt.subplots_adjust(hspace=1)
plt.subplots_adjust(wspace=1)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(441)
sns.barplot(data=lm10cr7,x='Season',y='Liga_Goals',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('Liga_Goals',size=10)
plt.xticks(rotation=45)

plt.subplot(442)
sns.barplot(data=lm10cr7,x='Season',y='Liga_Asts',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('Liga_Asts',size=10)
plt.xticks(rotation=45)


plt.subplot(443)
sns.barplot(data=lm10cr7,x='Season',y='Liga_Aps',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('Liga_Aps',size=10)
plt.xticks(rotation=45)


plt.subplot(444)
sns.barplot(data=lm10cr7,x='Season',y='Mins/Liga_Goals',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('Mins/Liga_Goals',size=10)
plt.xticks(rotation=45)


plt.subplot(445)
sns.barplot(data=lm10cr7,x='Season',y='Mins/Liga_Asts',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('Mins/Liga_Asts',size=10)
plt.xticks(rotation=45)


plt.subplot(446)
sns.barplot(data=lm10cr7,x='Season',y='Mins/Liga_Aps',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('Mins/Liga_Aps',size=10)
plt.xticks(rotation=45) 



plt.subplot(447)
sns.barplot(data=lm10cr7,x='Season',y='CL_Goals',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('CL_Goals',size=10)
plt.xticks(rotation=45) 

plt.subplot(448)
sns.barplot(data=lm10cr7,x='Season',y='CL_Asts',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('CL_Asts',size=10)
plt.xticks(rotation=45)

plt.subplot(448)
sns.barplot(data=lm10cr7,x='Season',y='CL_Aps',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('CL_Aps',size=10)
plt.xticks(rotation=45)

plt.subplot(448)
sns.barplot(data=lm10cr7,x='Season',y='Mins/CL_Goals',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('Mins/CL_Goals',size=10)
plt.xticks(rotation=45)

plt.subplot(448)
sns.barplot(data=lm10cr7,x='Season',y='Mins/CL_Asts',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('Mins/CL_Asts',size=10)
plt.xticks(rotation=45)


plt.subplot(448)
sns.barplot(data=lm10cr7,x='Season',y='Mins/CL_Aps',hue='Player',palette={'Messi':'red','Ronaldo':'black'})
plt.title('Mins/CL_Aps',size=10)
plt.xticks(rotation=45) 


plt.subplots_adjust(hspace=0.4)
plt.subplots_adjust(wspace=0.6)
```
